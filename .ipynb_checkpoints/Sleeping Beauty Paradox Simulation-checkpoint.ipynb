```python
import random

class SleepingBeauty:
    
    def __init__(self, chance_of_choosing_heads):
        self.chance_of_choosing_heads = chance_of_choosing_heads
        self.score = 0
     
    #Returns True IF Sleeping Beauty choose heads
    def choose_heads_or_tails(self):
        x = float(self.chance_of_choosing_heads)
        return random.random() < x
    
class SleepingBeautyParadox:
    
    def __init__(self, sleeping_beauty):
        self.sleeping_beauty = sleeping_beauty
        
    #Returns True IF Sleeping Beauty makes the correct predictions
    def run_experiement(self):
        
        coin_flip = random.random() < .5
        
        if coin_flip:
            return self.sleeping_beauty.choose_heads_or_tails()
        else:
            mondays_guess = self.sleeping_beauty.choose_heads_or_tails()
            tuesdays_guess = self.sleeping_beauty.choose_heads_or_tails()
            
            return not (mondays_guess and tuesdays_guess)

number_of_simulations = 10000000
number_of_correct_guesses = 0
sleeping_beauty = SleepingBeauty("1")
sleeping_beauty_paradox = SleepingBeautyParadox(sleeping_beauty)
for i in range(number_of_simulations):
    
    if sleeping_beauty_paradox.run_experiement():
        number_of_correct_guesses += 1
        
print(number_of_correct_guesses / number_of_simulations)
```

```python
digits = list("1234567890")
population = []
decimal_places_of_probabilities = 25
half_of_parents_genes = decimal_places_of_probabilities / 2
number_of_generations = 500

for i in range(300):
    chance_of_choosing_heads = "."
    for i in range(decimal_places_of_probabilities):
        chance_of_choosing_heads += random.choice(digits)
    population.append(SleepingBeauty(chance_of_choosing_heads))
    
for generation in range(number_of_generations):
    
    print(generation)
    
    #Calculate the score of each sleeping beauty
    for sleeping_beauty in population:
        sleeping_beauty_paradox = SleepingBeautyParadox(sleeping_beauty)
        number_of_simulations = 200000
        number_of_correct_guesses = 0
        
        for i in range(number_of_simulations):
            if sleeping_beauty_paradox.run_experiement():
                number_of_correct_guesses += 1
                
        sleeping_beauty.score = number_of_correct_guesses / number_of_simulations
      
    #If final generation show the best sleeping beauties
    if generation == number_of_generations - 1:
        max_score = max([sleeping_beauty.score for sleeping_beauty in population])
        best_sleeping_beauties = [sleeping_beauty for sleeping_beauty in population if sleeping_beauty.score == max_score]
        z = set()
        for best_sleeping_beauty in best_sleeping_beauties:
            z.add(best_sleeping_beauty.chance_of_choosing_heads)
        print(z)
        break
    
    #Create the next generation of sleeping beauties
    new_population = []
    total_score = sum([sleeping_beauty.score for sleeping_beauty in population])
    for i in range(0, len(population), 2):
        
        #Select Father
        previous_fitness = 0
        roulette = random.random()
        father = population[0]
        for sleeping_beauty in population:
            
            fitness_of_sleeping_beauty = (sleeping_beauty.score / total_score) + previous_fitness
            if roulette < fitness_of_sleeping_beauty:
                father = sleeping_beauty
                break
                
            previous_fitness = fitness_of_sleeping_beauty
        
        #Select Mother
        previous_fitness = 0
        roulette = random.random()
        mother = population[0]
        for sleeping_beauty in population:
            
            fitness_of_sleeping_beauty = (sleeping_beauty.score / total_score) + previous_fitness
            if roulette < fitness_of_sleeping_beauty:
                mother = sleeping_beauty
                break
                
            previous_fitness = fitness_of_sleeping_beauty
        
        #Add the parents or their children to the new population
        if random.random() >= .6:
            new_population.append(father)
            new_population.append(mother)
        else:
            x = random.randint(0, half_of_parents_genes-1)
            y = random.randint(half_of_parents_genes, decimal_places_of_probabilities-1)
            chance_of_choosing_heads_1 = "."
            chance_of_choosing_heads_2 = "."
            for i in range(decimal_places_of_probabilities):
                
                # Double crossover with a mutation
                if random.random < .2:
                    chance_of_choosing_heads_1 += random.choice(digits)
                    chance_of_choosing_heads_2 += random.choice(digits)
                elif i <= x or i > y:
                    chance_of_choosing_heads_1 += father.chance_of_choosing_heads[i]
                    chance_of_choosing_heads_2 += mother.chance_of_choosing_heads[i]
                else:
                    chance_of_choosing_heads_1 += mother.chance_of_choosing_heads[i]
                    chance_of_choosing_heads_2 += father.chance_of_choosing_heads[i]
                    
            new_population.append(SleepingBeauty(chance_of_choosing_heads_1))
            new_population.append(SleepingBeauty(chance_of_choosing_heads_2))
    
    population = new_population
```
